```python
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
#data = pd.read_sql_query('SELECT * FROM loan', engine)
#data.head()
```

```python
# How many distinct (different) actors' last names are there?
query = "SELECT count(distinct last_name)  FROM sakila.actor" 
data = pd.read_sql_query(query, engine)
data.head()
```

```python
# Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
query = 'SELECT *, \
CASE \
WHEN weekday(rental_date)  = 6 then "weekend" \
WHEN weekday(rental_date)  = 7 then "weekend" \
ELSE "workday" \
END AS "rental_date",rental_date AS "day_type" \
FROM sakila.rental;'

data = pd.read_sql_query(query, engine)
data.head()
```

```python
# Get all films with ARMAGEDDON in the title

query = "SELECT *  FROM sakila.film where title like '%%ARMAGEDDON%%';"
data = pd.read_sql_query(query, engine)
data.head()
```

```python
# Get 10 the longest film
query = 'SELECT title, length  FROM sakila.film ORDER BY LENGTH desc LIMIT 10;'
data = pd.read_sql_query(query, engine)
data
```

```python
# How many films include Behind the Scenes content?
query = 'SELECT count(film_id)  FROM sakila.film where special_features <> " ";'
data = pd.read_sql_query(query, engine)
data
```

```python
# Which kind of movies (rating) have a mean duration of more than two hours?
query = "SELECT rating FROM film  GROUP BY rating having avg(length)>120;"
data = pd.read_sql_query(query, engine)
data
```

```python
# Rank films by length (filter out the rows that have nulls or 0s in length column). 
#In your output, only select the columns title, length, and the rank.
query = 'SELECT title, length, rank() over(order by length desc) as ranking  FROM film having length <> 0 | " ";'
data = pd.read_sql_query(query, engine)
data
```
